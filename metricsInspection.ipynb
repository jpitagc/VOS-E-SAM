```python
import pandas as pd
import sys
```

```python
data = pd.read_csv('./result/all_tests.csv')
data
```

```python
sorted_16 = data[data.Test.str.contains('16')].sort_values(by = 'J&F-Mean',ascending=False)
sorted_16
```

```python
sorted_17 = data[data.Test.str.contains('17')].sort_values(by = 'J&F-Mean',ascending=False)
sorted_17
```

```python
f_16 = pd.read_csv(f'./result/{sorted_16.iloc[0].Test}/whole_metrics.csv')
s_16 = pd.read_csv(f'./result/{sorted_16.iloc[1].Test}/whole_metrics.csv')
t_16 = pd.read_csv(f'./result/{sorted_16.iloc[2].Test}/whole_metrics.csv')

cf_16 = f_16[~f_16.Video.str.contains('_object')].iloc[:,:-2].add_suffix('_X')
cs_16 = s_16[~s_16.Video.str.contains('_object')].iloc[:,1:-2].add_suffix('_RHQ')
ct_16 = t_16[~t_16.Video.str.contains('_object')].iloc[:,1:-2].add_suffix('_R')

concatenated_df = pd.concat([cf_16, cs_16, ct_16], axis=1)
num_cols = len(concatenated_df.columns)
intercalated_cols = cf_16.columns.to_list()[0:1] + [col for pair in zip(cf_16.columns.to_list()[1:], cs_16.columns, ct_16.columns) for col in pair]
concatenated_df = concatenated_df[intercalated_cols]
concatenated_df
```

```python
f_17 = pd.read_csv(f'./result/{sorted_17.iloc[0].Test}/whole_metrics.csv')
s_17 = pd.read_csv(f'./result/{sorted_17.iloc[1].Test}/whole_metrics.csv')
t_17 = pd.read_csv(f'./result/{sorted_17.iloc[2].Test}/whole_metrics.csv')

cf_17 = f_17[~f_17.Video.str.contains('_object')].iloc[:,:-2].add_suffix('_X')
cs_17 = s_17[~s_17.Video.str.contains('_object')].iloc[:,1:-2].add_suffix('_R')
ct_17 = t_17[~t_17.Video.str.contains('_object')].iloc[:,1:-2].add_suffix('_RHQ')

concatenated_df = pd.concat([cf_17, cs_17, ct_17], axis=1)
num_cols = len(concatenated_df.columns)
intercalated_cols = cf_17.columns.to_list()[0:1] + [col for pair in zip(cf_17.columns.to_list()[1:], cs_17.columns, ct_17.columns) for col in pair]
concatenated_df = concatenated_df[intercalated_cols]
concatenated_df
```
